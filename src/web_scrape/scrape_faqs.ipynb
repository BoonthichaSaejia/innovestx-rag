```python
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
```

```python
url = 'https://www.innovestx.co.th/faqs'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
}
response = requests.get(url,headers=headers)
page_html = response.content
data = soup(page_html, 'html.parser')
link_elements = data.find_all('a', class_='link_arrow purple_main')
links = [link.get('href') for link in link_elements]
links.append(url)
```

```python
faq_df = pd.DataFrame(columns=['text', 'context','topic','subtopic'])
for link in links:
    response = requests.get(link,headers=headers)
    page_html = response.content
    data = soup(page_html, 'html.parser')
    faqs = data.find_all('div', class_='accordion_tab')
    questions = []
    answers   = []
    topics = [] 
    subtopics = []
    for tab in faqs:
        text = 'กขจนกดเ'
        question_div = tab.find('div', class_='accordion_tab_label font_medium s_20')
        answer_div = tab.find('div', class_='accordion_tab_content')

        if question_div and answer_div:
            question = question_div.get_text(strip=True)
            answer_parts = []
            for elem in answer_div.descendants:
                if elem.name == 'a' and 'href' in elem.attrs:
                    href = elem['href']
                    if not href.startswith('http'):
                        href = 'https://www.innovestx.co.th' + href
                    answer_parts.append(elem.get_text(strip=True) + ' ' + href)
                    text = elem.get_text(strip=True)
                elif elem.name not in ['p','span','div','strong','br','sup'] and elem.get_text(strip=True) != text:
                    print(elem)
                    answer_parts.append(elem.get_text(strip=True))
            
            # Remove empty strings and join answer parts with a space
            answer = ' '.join(filter(None, answer_parts))
            questions.append(question)
            answers.append(answer)
            topics.append('faqs')
            subtopics.append(link.split('/')[-1])
    tmp_df = pd.DataFrame({'text': questions, 'context': answers, 'topic': topics, 'subtopic': subtopics})
    faq_df = pd.concat([faq_df, tmp_df], ignore_index=True)
    print(len(faq_df))


    
```
